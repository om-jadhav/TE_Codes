```python
text = """Data Science is an interdisciplinary field that uses scientifc methods, processes, algorithms and systems to extract knowledge from structured and unstructured Data."""
print(text)
```

```python
import nltk
import string
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import word_tokenize,sent_tokenize
```

```python
sentences = sent_tokenize(text)
print(sentences)
```

```python
words = word_tokenize(text)
print(words)
```

```python
words_no_punct = [word for word in words if word.isalnum()]
print(words_no_punct)
```

```python
from nltk.corpus import stopwords
```

```python
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words_no_punct if word.lower() not in stop_words]
print(filtered_words)
```

```python
from nltk import pos_tag
```

```python
nltk.download('averaged_perceptron_tagger_eng')
```

```python
pos_tags = pos_tag(filtered_words)
print(pos_tags)
```

```python
from nltk.stem import PorterStemmer,WordNetLemmatizer
```

```python
ps = PorterStemmer()
stemmed_words = [ps.stem(word) for word in filtered_words]
print(stemmed_words)
```

```python
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]
print(lemmatized_words)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
documents = [
    "Data Science is used for extracting knowledge.",
    "Machine learning is a part of Data Science.",
    "Big Data analytics helps in decision making."
]
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(documents)
print("Vocabulary:", vectorizer.get_feature_names())
print("TF-IDF Matrix:\n", tfidf_matrix.toarray())
```
