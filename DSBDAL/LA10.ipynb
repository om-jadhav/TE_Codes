```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("iris")
df.head()
```

```python
print(df.dtypes)
```

```python
df.hist(bins=15, figsize=(8,6))
plt.suptitle("Histogram of Iris Features")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(data=df)
plt.title("Boxplot of Iris Features")
plt.show()
```
